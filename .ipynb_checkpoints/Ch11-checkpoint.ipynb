```python
from arch import arch_model
```

```python
msft= pd.read_csv('../../datasets/Ch11/msft.csv',
                 index_col='Date',
                 parse_dates=True)
```

```python
msft
```

```python
msft['returns']=100*msft.pct_change()
msft.dropna(inplace=True,how='any')
```

```python
msft.plot(subplots=True,
         title='Microsoft Daily closing Price and Daily Returns')
```

```python
idx=round(len(msft)*0.90)
train=msft.returns[:idx]
test=msft.returns[idx:]
print(f'Train:{train.shape}')
print(f'Train:{test.shape}')
```

```python
model =arch_model(train, p=1,q=1,
                 mean='Constant',vol='GARCH',
                 dist='normal')
results=model.fit(update_freq=5)
```

```python
results.summary()
```

```python
results.pvalues
```

```python
results.plot()
plt.show()
```

```python
results.std_resid.hist(bins=20)
plt.title('Standardized Residuals')
```

```python
from statsmodels.stats.diagnostic import acorr_ljungbox
from statsmodels.stats.diagnostic import acorr_ljungbox
acorr_ljungbox(results.std_resid,
               lags=10,
               return_df=True)['lb_pvalue']
```

```python
msft_forecast = results.forecast(horizon=test.shape[0],method='simulation')
```

```python
def plot_forecast(model,start,train,test):
    forecast =pd.DataFrame(model.forecast(test.shape[0]),
                          index=test.index)
    ax = train.loc[start:].plot(style='--')
    test.plot(ax=ax)
    forecast.plot(ax=ax, style='-.')
    ax.legend(['orig_train','orig_test','forecast'])
    plt.show()
```

```python
forecast =msft_forecast.variance[-1:]
```

```python
forecast=pd.DataFrame(forecast.transpose(),index=test.index)
```

```python
ax = train.loc['2019-11':].plot(style='--')
test.plot(ax=ax)
forecast.plot(ax=ax, style='-.')
ax.legend(['orig_train','orig_test','forecast'])
plt.show()
```
